```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.spatial.distance import cdist
from sklearn.datasets import make_blobs
```

```python
X_train, true_labels = make_blobs(n_samples=500, centers= 5 , random_state=42, )
```

```python
plt.figure(figsize= (8,6))
for i in np.unique(true_labels):
    plt.scatter(X_train[true_labels == i , 0] , X_train[true_labels == i , 1] , label = i)
plt.legend()
plt.show()
```

```python
def calculate_distortion(x, centroids):
  m = len(x)
  return (1 / m) * sum(np.min(cdist(x, centroids ,'euclidean'), axis = 1)**2)
```

```python
def kmeans(x, k, no_of_iterations):
    ''' 
    K-means is a greedy algorithm that minimizes the squared distance of points from their respective cluster means
    It performs hard clustering, that is, each point is assigned to only one cluster.
    '''
    points_array = []
    centroids_array = []
    distorion = []
    for i in range(10):
      # K is chosen as the number of clusters we wish to cluster our data into

      # Randomly choose centroid for each cluster :
      # 1- Get a random index in range of X
      idx = np.random.choice(len(x), k, replace=False)   
      # 2- Get the corresponding data to these indices
      centroids = x[idx, :] 
      

      # finding the distance between centroids and all the data points
      distances = cdist(x, centroids ,'euclidean') 
      
      # Cluster assignment : Centroid with the minimum Distance
      points = np.array([np.argmin(i) for i in distances]) 
      
      for _ in range(no_of_iterations):
        # Repeat: Cluster Assignment & Centroid Adjustment

          centroids = []
          for idx in range(k):

              # Cluster Assignment:  Assign data to the cluster whose centroid is closest
              temp_cent = x[points == idx] 

              # Centroid Adjustment:  Move each centroid to the mean of data assigned to its cluster
              centroids.append(temp_cent.mean(axis = 0))
  
          
          centroids = np.vstack(centroids) 
          
          distances = cdist(x, centroids ,'euclidean')
          points = np.array([np.argmin(i) for i in distances])
        
      points_array.append(points)
      centroids_array.append(centroids)   
      distorion.append(calculate_distortion(x, centroids))
    
    minimum_distortion = np.argmin(distorion)
    return points_array[minimum_distortion], centroids_array[minimum_distortion]
```

```python
labels, centroids = kmeans(X_train, 5, 1000)
```

```python
plt.figure(figsize= (8,6))
for i in np.unique(labels):
    plt.scatter(X_train[labels == i , 0] , X_train[labels == i , 1] , label = i)
plt.scatter(centroids[:,0], centroids[:, 1], c = 'k', marker = 'x')
plt.legend()
plt.show()
```

```python
def Elbow_method(x, range_of_k):
  distortion = []
  inertias = []
  m = len(x)
  for k in range(2, range_of_k + 1):
    labels, centroids = kmeans(x, k, 300)
    distortion.append((1 / m) * sum(np.min(cdist(x, centroids ,'euclidean'), axis = 1)**2))
    
  
  plt.style.use('fivethirtyeight')
  plt.figure(figsize= (8, 6))
  plt.plot(np.arange(2, range_of_k + 1), distortion, 'bx-')
  plt.xlabel('Values of K')
  plt.ylabel('Distortion')
  plt.title('The Elbow Method using Distortion')
  plt.show()

```

```python
Elbow_method(X_train, 10)
```
